```python
import pandas as pd

# Load the Excel file
file_path = r'C:\Users\user\OneDrive\Υπολογιστής\Book2.xlsx'
sheet1 = 'Sheet1'
sheet2 = 'Sheet4'

# Read data from Sheet1, assuming the data starts from row 2 to row 289 (Excel index)
df = pd.read_excel(file_path, sheet_name=sheet1)

# Apply rolling median with a window of 12 values to each column A to H
smoothed_data = df.rolling(window=12, min_periods=1).median()

# Create an empty DataFrame with the same index as smoothed_data but with the necessary offset
output_df = pd.DataFrame(index=range(288), columns=smoothed_data.columns)

# Fill the DataFrame with the smoothed data
output_df.iloc[:smoothed_data.shape[0], :] = smoothed_data.values

# Write the smoothed data to a new sheet (Sheet4) starting from row 3
with pd.ExcelWriter(file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    output_df.to_excel(writer, sheet_name=sheet2, index=False, startrow=2)
```
